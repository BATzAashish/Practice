```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df= pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,s=200,alpha=0.2)
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.displot(data=df,x='salary',bins=100)
```

```python

sns.displot(data=df,x='salary',bins=20,color='red',edgecolor='blue',linewidth=4,ls='--')
```

```python
sns.histplot(data=df,x='salary',kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```
